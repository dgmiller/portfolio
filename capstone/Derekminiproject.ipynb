```python
geo = []
for coords in C.json_data:
    try:
        if coords['place'] is not None:
            geo.append(ut.np.mean(coords['place']['bounding_box']['coordinates'][0],axis=0))
    except:
        continue
geo = ut.np.row_stack(geo)
```

```python
from mpl_toolkits.basemap import Basemap
%matplotlib inline
print geo
```

```python
ut.plt.figure(figsize=(15,10))
my_map = Basemap(projection='robin',lon_0=0,lat_0=0,resolution='c',area_thresh=1000)
my_map.drawcoastlines(color='grey')
my_map.drawcountries(color='grey')
my_map.drawstates(color='grey')
x,y = my_map(geo[:,0],geo[:,1])
# scatter plot with constant bubble size
my_map.scatter(x,y,s=50,color='r',alpha=.4,lw=0)
ut.plt.show()
```

```python
ut.plt.figure(figsize=(15,10))
my_map = Basemap(projection='merc',lat_0=33.,lon_0=-125.,resolution='l',area_thresh=1000.,
                llcrnrlon=-130.,llcrnrlat=25,urcrnrlon=-65., urcrnrlat=50)
my_map.drawcoastlines(color='grey')
my_map.drawcountries(color='grey')
my_map.drawstates(color='grey')
#my_map.drawlsmask(land_color='white',ocean_color='blue')
x,y = my_map(geo[:,0],geo[:,1])
# scatter plot with constant bubble size
my_map.scatter(x,y,s=50,color='r',alpha=.4,lw=0)
ut.plt.show()
```
